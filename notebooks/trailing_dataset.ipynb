```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

"""sys.path.insert(0, './src')
from tools import *
from plotting_tools import *"""

#import processed datasets
path_to_data = '../_data/'

filtered_data = pd.read_csv(path_to_data+'filtered_data.csv')
master_boxscore = pd.read_csv(path_to_data+'master_boxscore.csv').sort_values(by='gmDate',ascending=False).drop(columns=['Unnamed: 0'])
master_OU = pd.read_csv(path_to_data+'master_OU.csv')
```

```python
game_data = filtered_data[['Date', 'opptAbbr', 'teamAbbr', 'Spread', 'Open_Line_OU', 'O/U',]]
game_data
```

```python
game_data_array = game_data.values.tolist()
game_data_array[0]
```

```python
row_list = []

for row in tqdm(game_data_array):  
    team_temp = master_boxscore.loc[(master_boxscore['gmDate'] <= row[0]) & (master_boxscore['teamAbbr'] == row[2])][team_columns][1:6]
    new_team_row = team_temp.mean()
    new_team_row['teamDayOff'] = team_temp.teamDayOff.sum()
    
    oppt_temp = master_boxscore.loc[(master_boxscore['gmDate'] <= row[0]) & (master_boxscore['opptAbbr'] == row[1])][oppt_columns][1:6]
    new_oppt_row = oppt_temp.mean()
    new_oppt_row['opptDayOff'] = oppt_temp.opptDayOff.sum()
    
    row_list.append(pd.concat([new_team_row, new_oppt_row]).to_dict())
```

```python
game_boxscore = pd.DataFrame(row_list)
game_boxscore.head()
```

```python
game_data = game_data.join(game_boxscore)
game_data
```
